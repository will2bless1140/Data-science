```python
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
nltk.download("punkt")
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

ps = PorterStemmer()
filtered_docs = []
for doc in data:
  tokens = word_tokenize(doc)
  tokens_lower = [w.lower() for w in tokens if w.isalnum()]
  tmp = ""
  for w in tokens_lower:
    if w not in stopwords:


      tmp += ps.stem(w) + " "
      print(w)
      print(tmp)
      print('---------------')
  filtered_docs.append(tmp)
print(filtered_docs)

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
x = vectorizer.fit_transform(filtered_docs)
print(x.todense())
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from yellowbrick.cluster import KElbowVisualizer

# sample data
xx, y = make_blobs(n_samples=366, centers=3, random_state=42)

# instantiate KMeans model and elbow visualizer
k = KMeans(init='k-means++', max_iter=100, n_init=1)
visualizer = KElbowVisualizer(k, k=(1,10))

# fit and
visualizer.fit(xx)

#visualize elbow plot
elbow_line = visualizer.ax.lines[0]
elbow_line.set_color('r')
visualizer.show()

optimal_k = visualizer.elbow_value_
# print optimal value of k
print("Optimal number of clusters: ", optimal_k)
```

```python
# using Kmeans
from sklearn.cluster import KMeans
k = 3
model = KMeans(n_clusters=k,init ='k-means++',max_iter=100,n_init=1)
model.fit(x)

print("cluster no. of input documents, in the order they recieved")
print(model.labels_)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
from scipy.sparse import csr_matrix

# generating sample data
x, y = make_blobs(n_samples=366, centers=3, cluster_std=0.6, random_state=42)

# check if x is a csr_matrix and convert it to array if necessary
if isinstance(x, csr_matrix):
    x = x.toarray()

# perform clustering
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=100, n_init=1)
kmeans.fit(x)
centroids = kmeans.cluster_centers_
inertia = kmeans.inertia_

# assigning labels to each data point
labels = kmeans.predict(x)

# list of colors for each cluster
colors = ['blue', 'green', 'orange']

# map labels to colors and plot the data points and centroids with category labels
for label in np.unique(labels):
    plt.scatter(x[labels == label, 0], x[labels == label, 1], c=colors[label], label='Cluster {}'.format(label+1))
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=500, linewidths=3, c=colors)
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=500, linewidths=3, color="red")
# Add title and legend to the plot
plt.title('Inertia Score: {:.2f}'.format(inertia))
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np
from scipy.sparse import csr_matrix

# generating sample data
x, y = make_blobs(n_samples=366, centers=3, cluster_std=0.6, random_state=42)

# check if x is a csr_matrix and convert it to array if necessary
if isinstance(x, csr_matrix):
    x = x.toarray()

# range of cluster numbers to use
range_of_clusters = range(2, 10)

# list to store average silhouette scores for each cluster number
silhouette_average = []

# loop through each cluster number and calculate average silhouette score
for no_clusters in range_of_clusters:
    # initialise kmeans
    kmeans = KMeans(n_clusters=no_clusters, init='k-means++', max_iter=100, n_init=1)
    kmeans.fit(x)
    # using predict to assign labels
    cluster_labels = kmeans.predict(x)

    # silhouette score
    silhouette_average.append(silhouette_score(x, cluster_labels))

# find the cluster number with the highest silhouette score
best_num_clusters = range_of_clusters[np.argmax(silhouette_average)]
best_silhouette_score = np.max(silhouette_average)

# print the results
print(f"Best cluster number: {best_num_clusters}")
print(f"Highest Silhouette score: {best_silhouette_score}")

# plot the silhouette scores for each cluster number
plt.plot(range_of_clusters, silhouette_average, 'rx--')
plt.xlabel('K number')
plt.ylabel('Silhouette score')
plt.title('Silhouette plot for KMeans clustering')
plt.show()
```
